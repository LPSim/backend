````python
import os
import json
target_folder = '../logs/obj_name_descs/'
files = os.listdir(target_folder)
files = [target_folder + file for file in files]
result = {}
for file in files:
    with open(file, 'r') as f:
        data = json.load(f)
        for type in data.keys():
            if type not in result.keys():
                result[type] = {}
            for name in data[type].keys():
                if name not in result[type].keys():
                    result[type][name] = {}
                for desc in data[type][name].keys():
                    result[type][name][desc] = 1
print(json.dumps(result, indent = 2, sort_keys = True))
````

````python
result_name = {}
for type in result.keys():
    for name in result[type].keys():
        result_name[f'{type}/{name}'] = ''
result_desc = {}
for type in result.keys():
    for name in result[type].keys():
        for desc in result[type][name].keys():
            result_desc[f'{type}/{name}/{desc}'] = ''
print(json.dumps(result_name, indent = 2, sort_keys = True))
print(json.dumps(result_desc, indent = 2, sort_keys = True))
````

```javascript
kapai_data = document.querySelectorAll('.kapai-data')

function getnext(node) {
  while (node.nextSibling.tagName != 'DIV') {
    node = node.nextSibling
  }
  return node.nextSibling
}

// deal charactor cards
charactor_data = {}
for (let i = 0; i < kapai_data.length; i ++ ) {
  if (kapai_data[i].parentNode.style.display == 'none') continue
  let node = kapai_data[i]
  let name = node.querySelector('a').innerText
  let skill_titles = node.querySelectorAll('.jiNeng-title')
  let skills = []
  for (let j = 0; j < skill_titles.length; j ++) {
    let skill = skill_titles[j]
    let skill_name = skill.innerText
    let skill_type = getnext(skill).innerText
    let skill_desc = getnext(getnext(skill)).innerText
    skills.push([skill_name, skill_type, skill_desc])
  }
  charactor_data[name] = skills
}

// other cards
other_data = {}
for (let i = 0; i < kapai_data.length; i ++ ) {
  if (kapai_data[i].parentNode.style.display != 'none') continue
  let node = kapai_data[i]
  let name = node.querySelector('a').innerText
  let title = node.querySelector('.jiNeng-title')
  other_data[name] = getnext(title).innerText
}
```
